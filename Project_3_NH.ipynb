# **Using IQR Technique for outliers**

```python
import numpy as np
```

```python
def iqr_technique(DFcolumn):
  Q1 = np.percentile(DFcolumn, 25)
  Q3 = np.percentile(DFcolumn, 75)
  IQR = Q3 - Q1
  lower_range = Q1 - (1.5 * IQR)
  upper_range = Q3 + (1.5 * IQR)
  return lower_range, upper_range
```

```python
lower_range, upper_range = iqr_technique(df['SalePrice'])
df = df[(df.SalePrice > lower_range) & (df.SalePrice <upper_range)]
```

# ***Pricing Trends Over Time***

```python
df['YrSold'] = pd.to_datetime(df['YrSold'], format='%Y')
df['YearSold'] = df['YrSold'].dt.year
```

```python
yearly_trends = df.groupby('YearSold')['SalePrice'].mean().reset_index()
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=yearly_trends, x='YearSold', y='SalePrice')
plt.title('Average Sale Price Over Years')
plt.xlabel('Year Sold')
plt.ylabel('Average Sale Price')
plt.grid(True)
plt.show()
```

```python
external_data = pd.DataFrame({
    'Year': [2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023],
    'InterestRate': [6.5, 6.0, 5.5, 5.0, 4.5, 4.0, 3.5, 3.0, 3.0, 3.5, 3.5, 4.0, 4.5, 4.5, 4.0, 3.5, 3.0, 3.0],
    'GDPGrowth': [2.7, 1.9, -0.1, -2.5, 2.6, 1.5, 2.2, 2.5, 2.4, 2.9, 1.6, 2.4, 2.9, 2.3, -3.5, 5.7, 2.0, 2.1],
    'InflationRate': [3.2, 2.8, 3.8, -0.4, 1.6, 3.2, 2.1, 1.5, 1.6, 0.1, 1.3, 2.1, 1.9, 2.3, 1.2, 1.7, 7.0, 4.9]
})
```

# ***Investigate Market Influences***

```python
merged_data = pd.merge(yearly_trends, external_data, left_on='YearSold', right_on='Year')
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=merged_data, x='Year', y='SalePrice', label='Average Sale Price')
sns.lineplot(data=merged_data, x='Year', y='InterestRate', label='Interest Rate', color='red')
plt.title('Impact of Interest Rates on Average Sale Price')
plt.xlabel('Year')
plt.ylabel('Values')
plt.legend()
plt.grid(True)
plt.show()
```

```python
correlation_with_external = merged_data.corr()['SalePrice']
print(correlation_with_external)
```

```python
plt.figure(figsize=(12, 6))
sns.heatmap(merged_data.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Between House Prices and Market Influences')
plt.show()
```
